# DATASET TB/U

Pembuatan Dataset Tinggi Badan Balita Menurut Usia di Bawah 5 Tahun Berdasarkan Standar WHO
Sumber: https://www.who.int/tools/child-growth-standards/standards/weight-for-age
Menggabungkan dataset di bawah ini:
- Girls table- Length-for-age: Birth to 2 years (z-scores)
- Girls table- Height-for-age: 2 to 5 years (z-scores)
- Boys table- Length-for-age: Birth to 2 years (z-scores)
- Boys table- Height-for-age: 2 to 5 years (z-scores)

```python
# Membaca data tinggi badan menurut usia untuk balita laki-laki di bawah 2 tahun
df_tbu_male_1 = pd.read_excel(r"D:\Prediksi Status Gizi\Dataset\lhfa_boys_0-to-2-years_zscores.xlsx")
df_tbu_male_1.head()
```

```python
df_tbu_male_1.info()
```

```python
df_tbu_male_1.isnull().sum()
```

```python
df_tbu_male_1.duplicated().sum()
```

```python
# Membaca data tinggi badan menurut usia balita laki-laki untuk usia 2 sampai 5 tahun
df_tbu_male_2 = pd.read_excel(r"D:\Prediksi Status Gizi\Dataset\lhfa_boys_2-to-5-years_zscores.xlsx")
df_tbu_male_2.head()
```

```python
df_tbu_male_2.info()
```

```python
df_tbu_male_2.isnull().sum()
```

```python
df_tbu_male_2.duplicated().sum()
```

Pada data "lhfa_boys_2-to-5-years_zscores" baris dengan nilai 'Month' sama dengan 24 akan dihapus. Karena hal itu data yang digunakan untuk usia 24 bulan ada pada data "lhfa_boys_0-to-2-years_zscores" dengan pengukuran PB dilakukan dalam keadaaan anak telentang

```python
# Pada data "lhfa_boys_2-to-5-years_zscores" baris dengan nilai 'Month' sama dengan 24 akan dihapus. 
drop_month = df_tbu_male_2[df_tbu_male_2['Month'] == 24].index
df_tbu_male_2 = df_tbu_male_2.drop(drop_month)
```

```python
df_tbu_male_2.head()
```

```python
df_tbu_male= pd.concat([df_tbu_male_1, df_tbu_male_2], ignore_index=True)
df_tbu_male["Gender"] = "male"

df_tbu_male.head()
```

```python
df_tbu_male.info()
```

```python
print(df_tbu_male)
```

```python
# Membaca data tinggi badan menurut usia balita perempuan di bawah 2 tahun
df_tbu_female_1 = pd.read_excel(r"D:\Prediksi Status Gizi\Dataset\lhfa_girls_0-to-2-years_zscores.xlsx")
df_tbu_female_1.head()
```

```python
df_tbu_female_1.isnull().sum()
```

```python
df_tbu_female_1.duplicated().sum()
```

```python
# Membaca data tinggi badan menurut usia balita perempuan untuk usia 2 sampai 5 tahun
df_tbu_female_2 = pd.read_excel(r"D:\Prediksi Status Gizi\Dataset\lhfa_girls_2-to-5-years_zscores.xlsx")
df_tbu_female_2.head()
```

```python
df_tbu_female_2.isnull().sum()
```

```python
df_tbu_female_2.duplicated().sum()
```

```python
# Pada data "lhfa_girls_2-to-5-years_zscores" baris dengan nilai 'Month' sama dengan 24 akan dihapus
drop_month = df_tbu_female_2[df_tbu_female_2['Month'] == 24].index
df_tbu_female_2 = df_tbu_female_2.drop(drop_month)
```

```python
df_tbu_female_2.head()
```

```python
df_tbu_female= pd.concat([df_tbu_female_1, df_tbu_female_2], ignore_index=True)
df_tbu_female["Gender"] = "female"

df_tbu_female.head()
```

```python
df_tbu_female.info()
```

```python
# Menggabungkan data tinggi badan menurut usia balita laki laki dan perempuan
df_tb_u = pd.concat([df_tbu_male, df_tbu_female], ignore_index=True)
```
